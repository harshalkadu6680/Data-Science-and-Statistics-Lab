```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
np.arange(0,100,10)
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

# BAR CHART

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```

```python
x = np.arange(1, 11)                 # x values (1 to 10)
y = np.random.randint(1, 100, 10)    # random integers for y

# Create bar chart
plt.bar(x, y, color='skyblue', edgecolor='black')

# Add labels and title
plt.title("Random Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis (Random Values)")

# Show the chart
plt.show()
```

# scatter

```python
a=(1,2,4,5,6,8,9,3,4)
b=(10,30,35,79,33,44,87,56,55)

plt.scatter(a,b)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show
```
